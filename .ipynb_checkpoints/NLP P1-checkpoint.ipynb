```python
def get_unigram_counts(words):
    corpus = {}
    for word in words.split():
        if word not in corpus:
            corpus[word] = 1
        else:
            corpus[word] += 1
    return corpus        
```

```python
def get_bigram_counts(words):
    corpus = {}
    for prev_i, current_word in enumerate(words.split()[1:]):
        if (words[prev_i], current_word) not in corpus:
            corpus[(corpus[previous_i], current_word)] = 1
        else:
            corpus[(corpus[previous_i], current_word)] += 1
    return corpus
        
```
